```python
import json
import numpy as np
tweets = json.load(open("tweets_01-08-2021.json"))
```

```python
from wizard import *
%load_ext autoreload
%autoreload 2
from HashMap import *
```

```python
tweetMap = HashMap(10, .75)
for t in tweets:
    text = t['text']
    for w in text.split():
        w = w.lower() # Convert to lowercase
        if w in tweetMap:
            tweetMap[w] += 1
        else:
            tweetMap[w] = 1
```

```python
keys = tweetMap.keys() 
values = []
for key in keys:
    values.append(tweetMap[key])

printOrder = np.argsort(values)

for i in range(1, 1000):
    print(f'{i}: {keys[printOrder[-i]]}')

```
