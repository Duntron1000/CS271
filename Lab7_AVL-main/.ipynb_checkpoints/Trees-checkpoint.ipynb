```python
import numpy as np
import matplotlib.pyplot as plt

class TreeNode(object):
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.inorder_pos = 0
        self.height = 0
    
    def add(self, key):
        ret = self
        if key < self.key:
            if self.left:
                self.left = self.left.add(key)
            else:
                self.left = TreeNode(key)
        elif key > self.key:
            if self.right:
                self.right = self.right.add(key)
            else:
                self.right = TreeNode(key)
                
        self.height = max(self.get_height(self.left), self.get_height(self.right)) + 1
        
        if self.get_height(self.right) == self.get_height(self.left) - 2:
            if self.get_height(self.left.right) > self.get_height(self.left.left):
                self.left = self.left.rotate_ccw()
            ret = self.rotate_cw()
        elif self.get_height(self.left) == self.get_height(self.right) - 2:
            if self.get_height(self.right.right) < self.get_height(self.right.left):
                self.right = self.right.rotate_cw()
            ret = self.rotate_ccw()

        return ret
    
    def rotate_ccw(self):
        new_root = self.right
        self.right = new_root.left
        new_root.left = self

        self.height = max(self.get_height(self.left), self.get_height(self.right)) + 1
        new_root.height = max(new_root.get_height(self.left), new_root.get_height(self.right)) + 1

        return new_root
    
    def rotate_cw(self):
        new_root = self.left
        self.left = new_root.right
        new_root.right = self

        self.height = max(self.get_height(self.left), self.get_height(self.right)) + 1
        new_root.height = max(new_root.get_height(self.left), new_root.get_height(self.right)) + 1

        return new_root
    
    def get_height(self, node):
        if node is None:
            return 0
        return node.height
    
    def inorder(self, num, key_list):
        """
        Parameters
        ----------
        num: list
            List of a single element which keeps 
            track of the number I'm at
        """
        if self.left:
            self.left.inorder(num, key_list)
        self.inorder_pos = num[0]
        key_list.append(self.key)
        num[0] += 1
        if self.right:
            self.right.inorder(num, key_list)
    
    def draw(self, y):
        x = self.inorder_pos
        plt.scatter([x], [y], 50, 'k')
        plt.text(x+0.2, y, "{}".format(self.key))
        plt.text(x+0.7, y, "({})".format(self.height), color='r')
        y_next = y-1
        if self.left:
            x_next = self.left.inorder_pos
            plt.plot([x, x_next], [y, y_next])
            self.left.draw(y_next)
        if self.right:
            x_next = self.right.inorder_pos
            plt.plot([x, x_next], [y, y_next])
            self.right.draw(y_next)
        
        
class BinaryTree(object):
    def __init__(self):
        self.root = None
    
    def inorder(self):
        key_list = []
        if self.root:
            self.root.inorder([0], key_list)
        return key_list
    
    def draw(self):
        self.inorder()
        if self.root:
            self.root.draw(0)
    
    def add(self, key):
        if self.root:
            self.root = self.root.add(key)
        else:
            self.root = TreeNode(key)

def make_tree():
    T = BinaryTree()
    for val in [10, 7, 16, 3, 9, 11, 20, 14, 17, 13, 12, 6, 5, 17, 15, 8, 4]:
        T.add(val)
    return T

T = make_tree()
plt.figure()
T.draw()
```
